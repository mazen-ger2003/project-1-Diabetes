# | Data visualizing

```python
# histogram for each one.
df.hist(bins=8,figsize=(8,8))
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Outcome',data=df)
plt.title('Distribution of Diabetes Outcome')
plt.show()
```

```python
plt.figure(figsize=(5,5))
df.groupby('Outcome')['Age'].median().plot(kind='bar')
plt.title('Median Age of diabetic and non-diabetic patients')
plt.show()
```

```python
plt.figure(figsize=(5,5))
df.groupby('Outcome')['BMI'].mean().plot(kind='bar')
plt.title('Average BMI of diabetic and non-diabetic patients')
plt.show()

plt.figure(figsize=(5,5))
df.groupby('Outcome')['Insulin'].mean().plot(kind='bar')
plt.title('Average insulin level of diabetic and non-diabetic patients')
plt.show()
```

```python
plt.figure(figsize=(5,5))
sns.countplot(x='Pregnancies', hue ='Outcome', data=df, palette='Set2')
plt.show()
```

From the above plots, we can infer that:

1-Diabetes has affected older people (more than 10 years older than non-diabetic)

2-BMI and insulin level of non-diabetic people are higher. This may be attribute to patients undergoing insulin therapy.

3-Pregnancies of more than 6 represent a higher chance of being affected with diabetes.
